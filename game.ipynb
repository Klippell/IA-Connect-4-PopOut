#### Importes

```python
import numpy as np
```

#### Configurações do Tabuleiro


```python
ROWS = 6
COLS = 7
EMPTY = 0
PLAYER1 = 1  # X
PLAYER2 = 2  # O

def create_board():
    """Cria um tabuleiro vazio."""
    return np.zeros((ROWS, COLS), dtype=int)

def drop_piece(board, col, piece):
    """Insere uma peça no topo da coluna (Drop move)."""
    for r in range(ROWS-1, -1, -1):
        if board[r][col] == EMPTY:
            board[r][col] = piece
            return True
    return False

def pop_piece(board, col, piece):
    """Remove uma peça da base (Pop move)."""
    if board[ROWS-1][col] == piece:
        for r in range(ROWS-1, 0, -1):
            board[r][col] = board[r-1][col]
        board[0][col] = EMPTY
        return True
    return False

def check_win(board, piece):
    """Verifica 4 em linha (Horizontal, Vertical, Diagonais)."""
    for c in range(COLS-3):
        for r in range(ROWS):
            if all(board[r][c+i] == piece for i in range(4)): return True
    for c in range(COLS):
        for r in range(ROWS-3):
            if all(board[r+i][c] == piece for i in range(4)): return True
    for c in range(COLS-3):
        for r in range(ROWS-3):
            if all(board[r+i][c+i] == piece for i in range(4)): return True
    for c in range(COLS-3):
        for r in range(3, ROWS):
            if all(board[r-i][c+i] == piece for i in range(4)): return True
    return False

def check_winner_after_move(board, player_who_moved):
    """Regra de vitória dupla do PopOut[cite: 21]."""
    p1_wins = check_win(board, PLAYER1)
    p2_wins = check_win(board, PLAYER2)
    if p1_wins and p2_wins:
        return player_who_moved
    if p1_wins: return PLAYER1
    if p2_wins: return PLAYER2
    return None

def print_board(board):
    """Interface visual do tabuleiro[cite: 26, 31]."""
    print("\n  0 1 2 3 4 5 6")
    print("-" * 15)
    for r in range(ROWS):
        row_str = "|"
        for c in range(COLS):
            char = "." if board[r][c] == EMPTY else ("X" if board[r][c] == PLAYER1 else "O")
            row_str += char + " "
        print(row_str + "|")
    print("-" * 15)

def get_human_move(board, player):
    """Lida com o input do jogador humano[cite: 30]."""
    p_char = "X" if player == PLAYER1 else "O"
    while True:
        try:
            move_str = input(f"Jogador {p_char} - Coluna (0-6) e Tipo (d/p) ou 'q' para sair: ").lower()
            if move_str == "q" or move_str == "sair":
                return False
            
            move_input = move_str.split()
            col = int(move_input[0])
            m_type = move_input[1]
            if m_type == "d":
                if drop_piece(board, col, player): return True
            elif m_type == "p":
                if pop_piece(board, col, player): return True
            print("❌ Movimento inválido!")
        except Exception:
            print("❌ Use o formato: '3 d' ou '2 p'")
```

#### Cenários de Jogo


```python
def play_game(mode):
    board = create_board()
    current_player = PLAYER1
    
    while True:
        print_board(board)
        
        # Lógica de quem joga
        if mode == 1: # Humano vs Humano
            move_made = get_human_move(board, current_player)
            if not move_made:
                print("🔴 Jogo cancelado!")
                break
        elif mode == 2: # Humano vs Computador
            if current_player == PLAYER1:
                move_made = get_human_move(board, current_player)
                if not move_made:
                    print("🔴 Jogo cancelado!")
                    break
            else:
                print("🤖 Computador (MCTS) a pensar...")
                # Aqui entrará a função col, m_type = mcts_search(...)
                print("⚠️ IA ainda não implementada!")
                break
        elif mode == 3: # Computador vs Computador
            print("⚠️ IA vs IA ainda não implementada!")
            break

        # Verificação de fim de jogo
        winner = check_winner_after_move(board, current_player)
        if winner:
            print_board(board)
            print(f"🎉 VENCEDOR: JOGADOR {('X' if winner == PLAYER1 else 'O')}")
            break
            
        current_player = PLAYER2 if current_player == PLAYER1 else PLAYER1
```

#### Menu Principal


```python
# --- Menu Principal ---

def main_menu():
    while True:
        print("\n=== PopOut AI Project 2026 ===")
        print("1. Iniciar Jogo (Humano vs Humano)")
        print("2. Sair")
        
        opcao = input("\nEscolha uma opção: ")
        
        if opcao == "1":
            play_game(1)
        elif opcao == "2":
            print("👋 Até à próxima!")
            break
        else:
            print("❌ Opção inválida!")

if __name__ == "__main__":
    main_menu()
```
